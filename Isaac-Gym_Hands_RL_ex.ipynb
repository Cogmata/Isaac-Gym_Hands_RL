# Isaac-Gym - Simulation Environements for RL training of robotic hands

**Warning!** : Preview version of Isaac-Gym - Mostly for test-demo purpose

## Installation   

**Prerequisites**

* Ubuntu 18.04 or 20.04.
* Python 3.6, 3.7 or 3.8.
* Minimum recommended NVIDIA driver version: Linux: 460.32

**Download Isaac-Gym package**  
**Install in a new conda environment:**


```
./create_conda_env_rlgpu.sh
conda activate rlgpu
conda remove --name rlgpu --all  
```
**Install External RL library (rl_games):**
```
pip install git+git://github.com/Denys88/rl_games
```
